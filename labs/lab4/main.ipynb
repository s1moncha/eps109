1.3

```python
# If the [0] represents the rows and [1] the columns, then the rows are until pi and the columns span 2*pi
SA = 0
R = 40
theta = 2*math.pi/360
phi = math.pi/180
midpoint_adj = np.linspace(theta/2, np.pi-theta/2, 180)

for i in midpoint_adj:
    for j in range(360):
        SA += R**2 * np.sin(i) * phi * theta

print(f"Integral: {SA}")
print(f"Exact: {4*np.pi*(R**2)}")
print(f"Difference: {np.round(SA - (4*np.pi*(R**2)),4)}")
```

```python
water = 0
R = 40
theta = 2*math.pi/360
phi = math.pi/180
midpoint_adj = np.linspace(theta/2, np.pi-theta/2, 180)
longitude_tracker = 0

for i in midpoint_adj:
    for j in range(360):
        if (H[longitude_tracker][j] < 0):
            water += R**2 * np.sin(i) * phi * theta
    longitude_tracker += 1

print(f"Proportion of Water: {np.round(water/SA*100,2)}%")
```

```python
volume_water = 0
R = 6.378e10
theta = 2*math.pi/360
phi = math.pi/180
midpoint_adj = np.linspace(theta/2, np.pi-theta/2, 180)
longitude_tracker = 0

for i in midpoint_adj:
    for j in range(360):
        if (H[longitude_tracker][j] < 0):
            volume_water += (R**2 * np.sin(i) * phi * theta) * (H[longitude_tracker][j] * -1)
    longitude_tracker += 1

print(f"Volume of Water: {volume_water}")
```
