```python
n = 11
data = np.zeros((n,n))
# (a) what is the effect of the following three lines
data[1,1] = 1
data[2,3] = 1
data[3,5] = 1
# (b) again what do the following two lines do?
data[5,1:n-1] = 1
data[n-2,:] = 1
print(data)
plt.imshow(data, interpolation='nearest')
plt.show()
```

(a) The effect is that we are setting these three coordinates on the grid to always be labeled as yellow
(b) The effect is a way to mark cells as true, but across a range. The syntax is as follows: data[row, column], the column is shorthand for a range and if you don;t set a value for that it assumes across the entirey.

```python
# Here we practise 'for' loops
n = 11
data = np.zeros((n,n))

# (a) What is the effect of this 'for' loop?
for i in range(0,n):
    data[i,3] = 1

# (b) And this 'for' loop?
col = (n-1) // 2

for i in range(0,n,2):
    data[i,col] = 1

print(data)
plt.imshow(data, interpolation='nearest')
plt.show()
```

(a) The effect is we are filling in the 3rd column element for each row.  
(b) The effect is we’re doing the same thing, but are using a step of 2, which essentially mimics an alternating row behavior.

```python
n = 11
data = np.zeros((n,n))

for i in range(n):
    data[i,i] = 1
    data[i, n - i - 1] = 1

plt.imshow(data, interpolation='nearest')
plt.show()
```

```python
n = 11
data = np.zeros((n,n))

for i in range(1, 10):
    for j in range(1, 10):
        data[i,j] = 1

for i in range(2, 9):
    for j in range(2, 9):
        data[i,j] = 0

plt.imshow(data, interpolation='nearest')
plt.show()
```

```python
n = 11
data = np.zeros((n,n))

for i in range(1, 6, 2):
    moves = 2*i + 1
    for j in range(moves):
        data[5-i+j, 5-i] = 1
        data[5-i, 5-i+j] = 1
        data[5+i-j,5+i] = 1
        data[5+i,5+i-j] = 1

plt.imshow(data, interpolation='nearest')
plt.show()
```

```python
import math

n = 501
data = np.zeros((n,n))
print(len(data))


for i in range(n):
    for j in range(n):
        if (math.sqrt((250-i)**2 + (250-j)**2) < 250):
            data[i,j] = 1


plt.imshow(data, interpolation='nearest')
plt.show()
```

```python
H = np.loadtxt('topography_180x360_grid.txt')
print(H)
print(H.shape)
```

```python
plt.imshow(H, interpolation='nearest')
plt.show()
```

```python
H2 = H

for i in range(110, 160):
    for j in range(80, 140):
        H2[j,i] = 10000

plt.imshow(H2, interpolation='nearest')
plt.show()
```

```python
H3 = H

for i in range(75, 150):
    for j in range(275, 325):
        H3[i,j] = -10000

plt.imshow(H3, interpolation='nearest')
plt.show()
```

Interestingly, enough. Upon changing the H2 and H3, these changes were reflected in the H, too. Even though we didn't explictly change the values. This has to do with the mutability of arrays.

```python
H = np.loadtxt('topography_180x360_grid.txt')
H4 = H

for i in range(180):
    for j in range(360):
        if (H4[i,j] > 2000):
            H4[i,j] = 10000
        elif (H4[i,j] < -2000):
            H4[i,j] = -10000


plt.imshow(H4, interpolation='nearest')
plt.show()
```

```python
H = np.loadtxt('topography_180x360_grid.txt')
H4 = H

for i in range(180):
    for j in range(360):
        if (H4[i,j] > 2000):
            H4[i,j] = 10000
        elif (H4[i,j] < -2000):
            H4[i,j] = -10000
        else:
            H4[i,j] = 0


plt.imshow(H4, interpolation='nearest')
plt.show()
```

```python
import numpy as np

H5 = np.copy(H4)

left = H5[:, :187]
right = H5[:, 188:]

H5 = np.concatenate((right, left), axis=1)
plt.imshow(H5, interpolation='nearest')
plt.show()
```
