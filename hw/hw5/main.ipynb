```python
import numpy as np
import matplotlib.pyplot as plt

def Mandelbrot(z, C):
    zNew = z**2 + C
    return zNew
```

PROBLEM 1.1

```python
def complexConvergence(C, max_iter=200, threshold=50):
    value = C
    for i in range(max_iter):
        value = Mandelbrot(value, complex(-0.4, 0.6))
        if abs(value) > threshold:
            return i
    return max_iter

# Grid
x = np.linspace(-3, 3, 2001)
y = np.linspace(-3, 3, 2001)

result = np.zeros((len(y), len(x)))

for i, yy in enumerate(y):
    for j, xx in enumerate(x):
        C = complex(xx, yy)
        result[i, j] = complexConvergence(C)

plt.figure(figsize=(8, 8))
plt.imshow(result, extent=[x.min(), x.max(), y.min(), y.max()],
           cmap="inferno", origin="lower", interpolation="nearest")
plt.xlabel("Real")
plt.ylabel("Imaginary")
plt.title("Julia Set")
plt.savefig("julia_set.png", dpi=300, bbox_inches="tight")  
plt.show()
```
