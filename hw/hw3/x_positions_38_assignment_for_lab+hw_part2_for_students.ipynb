Review part 1 where we calculated the standard deviation for the difference between one data set and the solution. First pick one dataset and calculate the standard deviation in the same way. Then, for every dataset, compute the standard deviation between the data and the solution. Store standard deviation in a Numpy vector 'stdDev'.

```python
stdDev = []

for file in filenames:
    data = np.loadtxt(file)
    std = np.sqrt(np.sum((solutions - data)**2)/(len(data) - 1))
    stdDev.append(std)
```

Using the append() function, construct a long list **diffAll** with nFiles*nNumbers elements that contains the deviation between the everyone's guesses and the solution. Use the command plt.hist() from part 1 to plot the probability function for the deviations overall. Adjust the number of bins and the plot interval with plt.xlim(xMin,xMax) to obtain a similar graph as shown in lecture.

```python
diffAll = []

for file in filenames:
    data = np.loadtxt(file)
    diff = solutions - data
    diffAll.append(diff)

diffAll = np.array(diffAll).T.flatten()
```

```python
plt.hist(diffAll, bins=80, density=True)
plt.xlabel('Difference')
plt.xlabel('Density')
plt.title('Plotting std of everyone\'s response')
plt.show()
```

Again construct a long deviations vector **diffAll** but throw away all points with absolute deviation larger than 0.15. Plot another histogram that should look same as before except for the outliers. 

```python
removeOutliers = diffAll[(diffAll < 0.15) & (diffAll > -0.15)]

plt.hist(removeOutliers, bins=29, density=True)
plt.xlabel('Difference')
plt.xlabel('Density')
plt.title('Plotting std of everyone\'s response (NO OUTLIERS)')
plt.show()
```

Compute and print the standard deviation, sigma, of all elements in the updated variable **diffAll**. You may assume the average of all points is zero. Store the results in the variable 'sigma' and print its value so that our grader can find it.

```python
sigma = np.sqrt(np.sum(diffAll**2) / len(diffAll))
print(f"Standard deviation of elements in diffAll {sigma}")
```

Replot the histogram again with 'plt.hist(diffAll, 25,density=True)' and overplot a Gaussian Bell curve with the standard deviation 'sigma' that you just computed. 

```python
from scipy.stats import norm

plt.hist(diffAll, bins=25, density=True)
x = np.linspace(min(diffAll), max(diffAll), 1000)
plt.plot(x, norm.pdf(x, 0, sigma), 'r-', linewidth=2)
plt.xlabel('Difference')
plt.xlabel('Density')
plt.title('Class difference from solution with Bell Curve Overlay')
plt.show()
```
