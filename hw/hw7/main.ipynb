```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.animation import FFMpegWriter
```

```python
N = 1000
x = np.linspace(0, 1000, N)
y = np.linspace(0, 1000, N)
X, Y = np.meshgrid(x, y)

# Initialize field (all zeros)
Z = np.zeros_like(X)

for i in range(300, 700):
    for j in range(150, 250):
        Z[j][i] = 100
        Z[j+600][i] = 100

for i in range(450, 550):
    for j in range(250, 850):
        Z[j][i] = 100



# --- Visualize ---
plt.figure(dpi=200)
plt.imshow(Z, origin='lower', extent=[0, 1000, 0, 1000])
plt.title('Letter I on 2D Grid')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
L = 500
N = 100
x = np.linspace(0, L, N)
y = np.linspace(0, L, N)
X, Y = np.meshgrid(x, y)

T_old = np.zeros((N, N))
T_new = np.zeros_like(T_old)

T_old[int(0.15*N):int(0.25*N), int(0.30*N):int(0.70*N)] = 100
T_old[int(0.75*N):int(0.85*N), int(0.30*N):int(0.70*N)] = 100
T_old[int(0.25*N):int(0.75*N), int(0.45*N):int(0.55*N)] = 100

tol = 1e-4
max_iter = 5000

plt.ion()
fig, ax = plt.subplots(dpi=200)
im = ax.imshow(T_old, origin='lower', cmap='hot', extent=[0, L, 0, L], vmin=0, vmax=100)
cbar = fig.colorbar(im, ax=ax, label='Temperature (C)')
ax.set_title('Heat Diffusion  "I" Shape Cooling')
ax.set_xlabel('X')
ax.set_ylabel('Y')

metadata = dict(title='Heat Diffusion - Letter I (Cooling)', artist='Matplotlib')
writer = FFMpegWriter(fps=15, metadata=metadata)

energies = []
iterations = []

with writer.saving(fig, "heat_I_cooling.mp4", dpi=200):
    for it in range(max_iter):
        for i in range(1, N-1):
            for j in range(1, N-1):
                T_new[i, j] = 0.25 * (
                    T_old[i+1, j] + T_old[i-1, j] +
                    T_old[i, j+1] + T_old[i, j-1]
                )

        T_new[:,  0] = 0
        T_new[:, -1] = 0
        T_new[0,  :] = 0
        T_new[-1, :] = 0

        total_energy = np.sum(T_new)
        energies.append(total_energy)
        iterations.append(it)

        if it % 25 == 0:
            im.set_data(T_new)
            ax.set_title(f'"I" Cooling  Iteration {it}\nTotal Energy = {total_energy:.2f}')
            plt.draw()
            plt.pause(0.01)
            writer.grab_frame()

        diff = np.max(np.abs(T_new - T_old))
        if diff < tol:
            print(f" Converged after {it+1} iterations (max diff = {diff:.2e})")
            im.set_data(T_new)
            ax.set_title(f'Final (Converged at Iter {it+1})\nTotal Energy = {total_energy:.2f}')
            writer.grab_frame()
            break

        T_old[:] = T_new

plt.ioff()
plt.close(fig)
print(" Saved animation as heat_I_cooling.mp4")

plt.figure(dpi=200);
plt.plot(iterations, energies, color='blue', lw=2)
plt.title('Total Heat Energy Over Time')
plt.xlabel('Iteration')
plt.ylabel('Total Energy (Sum of Temperatures)')
plt.grid(True)
plt.show()
```

```python
L = 500
N = 100
x = np.linspace(0, L, N)
y = np.linspace(0, L, N)
X, Y = np.meshgrid(x, y)

T_old = np.zeros((N, N))
T_new = np.zeros_like(T_old)
T_old[int(0.15*N):int(0.25*N), int(0.30*N):int(0.70*N)] = 100
T_old[int(0.75*N):int(0.85*N), int(0.30*N):int(0.70*N)] = 100
T_old[int(0.25*N):int(0.75*N), int(0.45*N):int(0.55*N)] = 100

tol = 1e-4
max_iter = 5000

plt.ion()
fig, ax = plt.subplots(dpi=200)
im = ax.imshow(T_old, origin='lower', cmap='hot', extent=[0, L, 0, L], vmin=0, vmax=100)
cbar = fig.colorbar(im, ax=ax, label='Temperature (C)')
ax.set_title('Heat Diffusion  "I" Shape (No Energy Loss)')
ax.set_xlabel('X')
ax.set_ylabel('Y')

metadata = dict(title='Heat Diffusion - Letter I (Constant Energy)', artist='Matplotlib')
writer = FFMpegWriter(fps=15, metadata=metadata)

energies, iterations = [], []

with writer.saving(fig, "heat_I_constant_energy.mp4", dpi=200):
    for it in range(max_iter):
        for i in range(1, N-1):
            for j in range(1, N-1):
                T_new[i, j] = 0.25 * (
                    T_old[i+1, j] + T_old[i-1, j] +
                    T_old[i, j+1] + T_old[i, j-1]
                )

        T = T_old

        for i in range(1, N-1):
            T_new[i, 0]   = 0.25 * (T[i, 0] + T[i+1, 0] + T[i-1, 0] + T[i, 1])
            T_new[i, -1]  = 0.25 * (T[i, -1] + T[i+1, -1] + T[i-1, -1] + T[i, -2])

        for j in range(1, N-1):
            T_new[0, j]   = 0.25 * (T[0, j] + T[1, j] + T[0, j-1] + T[0, j+1])
            T_new[-1, j]  = 0.25 * (T[-1, j] + T[-2, j] + T[-1, j-1] + T[-1, j+1])

        T_new[0, 0]       = 0.25 * (2.0*T[0, 0]    + T[1, 0]    + T[0, 1])
        T_new[0, -1]      = 0.25 * (2.0*T[0, -1]   + T[1, -1]   + T[0, -2])
        T_new[-1, 0]      = 0.25 * (2.0*T[-1, 0]   + T[-2, 0]   + T[-1, 1])
        T_new[-1, -1]     = 0.25 * (2.0*T[-1, -1]  + T[-2, -1]  + T[-1, -2])

        total_energy = np.sum(T_new)
        energies.append(total_energy)
        iterations.append(it)

        if it % 25 == 0:
            im.set_data(T_new)
            ax.set_title(f'Iteration {it}  Total Energy = {total_energy:.2f}')
            plt.draw()
            plt.pause(0.01)
            writer.grab_frame()

        diff = np.max(np.abs(T_new - T_old))
        if diff < tol:
            print(f" Converged after {it+1} iterations (max diff = {diff:.2e})")
            im.set_data(T_new)
            ax.set_title(f'Final (Converged at {it+1})  Total Energy = {total_energy:.2f}')
            writer.grab_frame()
            break

        T_old[:] = T_new

plt.ioff()
plt.close(fig)

plt.figure(dpi=200)
plt.plot(iterations, energies, color='green', lw=2)
plt.title('Total Heat Energy (No Leakage)')
plt.xlabel('Iteration')
plt.ylabel('Total Energy (Sum of Temperatures)')
plt.grid(True)
plt.show()
```
