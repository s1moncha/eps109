### a) Without using any functions except np.random.random() and np.zeros(), write code to fill a NumPy vector with 1000 random numbers that are uniformly distributed between -1 and +1.

### c) Imagine instead of 1000 elements you had 10$^6$ or 10$^{12}$ elements, what value for the magntide of the increase would your code print approximately?

```python
# a
v = np.zeros(1000)

for i in range(len(v)):
    r = np.random.random()
    x = np.random.random()
    if x > 0.5:
        v[i] = r
    else:
        v[i] = -r
```

```python
difference = 0
index = 0

for i in range(len(v) - 1):
    if (abs(v[i] - v[i+1]) > difference):
        difference = abs(v[i] - v[i+1])
        index = i

print(f"Index {i} with difference of {difference}")
```

## Question 4: Drawing ellipses

### Recreate the image shown below. The ellipses you draw do not need to match the figure perfectly, but they should resemble it in appearance.



### Remember how we drew filled circles berore. The border of a **circle** centered in $(x_0, y_0)$ satisfies the equation:  

$$(x - x_0)^2 + (y - y_0)^2 = r^2$$

### The border of the **ellipse** centered in $(x_0, y_0)$ satisfies the equation:  

$$\frac{(x - x_0)^2}{a^2} + \frac{(y - y_0)^2}{b^2} = 1$$

### where:  
### - 'a' is the semi-major axis (horizontal radius),  
### - 'b' is the semi-minor axis (vertical radius).  


```python
import math
```

```python
n = 501
data = np.zeros((n, n))

for i in range(n):
    for j in range(n):
        if (math.sqrt((100-i)**2 / 1.98**2 + (100-j)**2) < 50):
            data[i + 150,j + 20] = 1
            data[i + 150,j + 250] = 1

        if (math.sqrt((100-i)**2 + (100-j)**2 / 1.98**2) < 50):
            data[i + 290,j + 130] = 1
            data[i + 20,j + 130] = 1



plt.imshow(data, interpolation='nearest')
plt.show()
```

# Question 5: Decorating a tree
### The following code fills a *square* with 3000 uniformly distributed circles. We are asking you to modify this code to plot exactly 3000 circles that are uniformly distributed in a *triangle* as shown here:

```python
n=3000
counter = 0

while (counter != 3000):
    x = np.random.random()
    y = np.random.random()
    if (y < 0.5):
        if (y < x and x < 0.5):
            plt.plot(x,y,'ro',mfc='w')
            counter += 1
        elif (x > 0.5):
            if (y < 1-x):
                plt.plot(x,y,'ro',mfc='w')
                counter += 1  

plt.show()
```
